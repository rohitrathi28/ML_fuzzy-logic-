```python
# Triangular

import numpy as np
import matplotlib.pyplot as plt
import skfuzzy as fuzz

angles = np.linspace(-np.pi, np.pi, num=1000)

north = fuzz.trimf(angles, [-3*np.pi/4, -np.pi/2, -np.pi/4])
northeast =fuzz.trimf(angles, [-np.pi/2, -np.pi/4, 0])
east = fuzz.trimf(angles, [-np.pi/4, 0, np.pi/4])
southeast = fuzz.trimf(angles, [0, np.pi/4, np.pi/2])
south =fuzz.trimf(angles, [np.pi/4, np.pi/2, 3*np.pi/4])
southwest = fuzz.trimf(angles, [np.pi/2, 3*np.pi/4, np.pi])
west =  fuzz.trimf(angles, [3*np.pi/4, np.pi, 5*np.pi/4])
northwest = fuzz.trimf(angles, [-np.pi, -3*np.pi/4, -np.pi/2])

# Plotting
plt.figure(figsize=(10, 6))

plt.plot(angles, north, label='North')
plt.plot(angles, northeast, label='Northeast')
plt.plot(angles, east, label='East')
plt.plot(angles, southeast, label='Southeast')
plt.plot(angles, south, label='South')
plt.plot(angles, southwest, label='Southwest')
plt.plot(angles, west, label='West')
plt.plot(angles, northwest, label='Northwest')

plt.xlabel('Angle')
plt.ylabel('Membership Value')
plt.title('Cardinal Directions')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Gaussian and Triangular plots represented in one graph

import numpy as np
import matplotlib.pyplot as plt
import skfuzzy as fuzz

angles = np.linspace(-np.pi, np.pi, num=1000)

north = fuzz.gaussmf(angles, -np.pi/2, np.pi/4)
northeast =fuzz.gaussmf(angles, -np.pi/4, np.pi/4)
east =fuzz.gaussmf(angles, 0, np.pi/4)
southeast =fuzz.gaussmf(angles, np.pi/4, np.pi/4)
south =fuzz.gaussmf(angles, np.pi/2, np.pi/4)
southwest =fuzz.gaussmf(angles, 3*np.pi/4, np.pi/4)
west =fuzz.gaussmf(angles, np.pi, np.pi/4)
northwest =fuzz.gaussmf(angles, -3*np.pi/4, np.pi/4)

north_tri = fuzz.trimf(angles, [-3*np.pi/4, -np.pi/2, -np.pi/4])
northeast_tri =fuzz.trimf(angles, [-np.pi/2, -np.pi/4, 0])
east_tri = fuzz.trimf(angles, [-np.pi/4, 0, np.pi/4])
southeast_tri = fuzz.trimf(angles, [0, np.pi/4, np.pi/2])
south_tri =fuzz.trimf(angles, [np.pi/4, np.pi/2, 3*np.pi/4])
southwest_tri = fuzz.trimf(angles, [np.pi/2, 3*np.pi/4, np.pi])
west_tri =  fuzz.trimf(angles, [3*np.pi/4, np.pi, 5*np.pi/4])
northwest_tri = fuzz.trimf(angles, [-np.pi, -3*np.pi/4, -np.pi/2])

# Plotting
plt.figure(figsize=(10, 6))

plt.plot(angles, north, label='North (Gaussian)')
plt.plot(angles, northeast, label='Northeast (Gaussian)')
plt.plot(angles, east, label='East (Gaussian)')
plt.plot(angles, southeast, label='Southeast (Gaussian)')
plt.plot(angles, south, label='South (Gaussian)')
plt.plot(angles, southwest, label='Southwest (Gaussian)')
plt.plot(angles, west, label='West (Gaussian)')
plt.plot(angles, northwest, label='Northwest (Gaussian)')

plt.plot(angles, north_tri, '--', label='North (Triangular)', alpha=0.5)
plt.plot(angles, northeast_tri, '--', label='Northeast (Triangular)', alpha=0.5)
plt.plot(angles, east_tri, '--', label='East (Triangular)', alpha=0.5)
plt.plot(angles, southeast_tri, '--', label='Southeast (Triangular)', alpha=0.5)
plt.plot(angles, south_tri, '--', label='South (Triangular)', alpha=0.5)
plt.plot(angles, southwest_tri, '--', label='Southwest (Triangular)', alpha=0.5)
plt.plot(angles, west_tri, '--', label='West (Triangular)', alpha=0.5)
plt.plot(angles, northwest_tri, '--', label='Northwest (Triangular)', alpha=0.5)

plt.xlabel('Angle')
plt.ylabel('Membership Value')
plt.title('Cardinal Directions')
plt.legend()
plt.grid(True)
plt.show()
```
